Find the rainfall dataset that is to be considered for the exercise.. This data contains region(district) wise rainfall across India.

1. Import the data into Python environment as a Pandas DataFrame.
2. Check for missing values, if any and drop the corresponding rows.
3. Find the district that gets the highest annual rainfall.
4. Display the top 5 states that get the highest annual rainfall.
5. Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.
6. Display the state-wise mean rainfall for all the months using a pivot table.
7. Display the count of districts in each state.
8. For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall.

```python
#Import the data into Python environment as a Pandas DataFrame.

import pandas as pd

rain_data = pd.read_csv('/content/sample_data/rainfall_Data.csv')
rain_data.head(5)
```

```python
#Check for missing values, if any and drop the corresponding rows.
rain_data.info() #no missing values
```

```python
rain_data.describe()
```

```python
#Find the district that gets the highest annual rainfall.
#highest_rainfall = rain_data["ANNUAL"].max()

i = rain_data["ANNUAL"].idxmax()
print(rain_data.loc[[i]])
print(rain_data.at[i,'DISTRICT'])
```

```python
#Display the top 5 states that get the highest annual rainfall.

rain_data.sort_values('ANNUAL', ascending=False).head(5)
```

```python
#Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.
rain_data.drop(columns=['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'], inplace=True)
rain_data.head()
```

```python
#Display the state-wise mean rainfall for all the months using a pivot table.

rain_data_annual_mean = rain_data.groupby('STATE_UT_NAME')[['ANNUAL']].mean()
rain_data_annual_mean
```

```python
#Display the state-wise mean rainfall for all the months using a pivot table.
import numpy as np

pd.pivot_table(rain_data, index ='STATE_UT_NAME',
               values=['JAN', 'FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'],
               aggfunc=np.mean, sort = False)
```

```python
#Display the count of districts in each state.

rain_data.groupby('STATE_UT_NAME')[['DISTRICT']].count()
```

```python
#For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall.

id = rain_data.groupby(['STATE_UT_NAME'])[['MAY']].idxmax().values

for i in id:
  print(rain_data.loc[i][['STATE_UT_NAME', 'DISTRICT','MAY']])
```
